```python
# Symbolic Math
# Define symbolic variables
x, y = var('x, y')

# Symbolic expressions
f = x^2 + 2*x - 1
print(f)

# Differentiation
print(diff(f, x))

# Integration
print(integrate(f, x))

# Solving equations
print(solve(x^2 - 4 == 0, x, algorithm='sympy'))
```

```python
# Number Theory
# Prime factorization
print(factor(12345))

# Greatest Common Divisor (GCD)
print(gcd(12, 18))

# Least Common Multiple (LCM)
print(lcm(4, 6))

# Check if a number is prime
print(is_prime(97))
```

```python
# Plotting
# Define a function
f(x) = sin(x)

# Plotting a function
p = plot(f(x), (x, -pi, pi))
p.show()

# Parametric plot
t = var('t')
parametric_plot([cos(t), sin(t)], (t, 0, 2*pi)).show()
```

```python
# Linear Algebra
# Define a matrix
M = Matrix([[1, 2], [3, 4]])
print(M)

# Determinant
print(M.det())

# Inverse
print(M.inverse())
```
